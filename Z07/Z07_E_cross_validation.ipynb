# Zadanie
Proszę nauczyć wszytkie poznane modele regresji.
Który z nich działa najlepiej?

```python
auto_target = data["price"]
auto_data = data.drop(["price"],axis=1)
```

```python
auto_data.head()
```

```python
auto_target.head()
```

```python
y=auto_target
X=auto_data
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.pipeline import make_pipeline
from sklearn import model_selection
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
kfold = model_selection.KFold(n_splits=5)
```

```python
X_train, X_test, y_train, y_test = train_test_split(auto_data, auto_target, test_size=0.25, random_state=42)
```

```python
models = []
```

```python
grid_eln = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), ElasticNet(alpha=1)),
                    param_grid={'polynomialfeatures__degree': [i for i in range(5)],
                    'elasticnet__alpha': [10**i for i in range(-10,10)]},
                    cv=kfold,
                    refit=True)
```

```python
grid_ls = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), Lasso(alpha=1)),
                    param_grid={'polynomialfeatures__degree': [i for i in range(5)],
                    'lasso__alpha': [10**i for i in range(-10,10)]},
                    cv=kfold,
                    refit=True)
```

```python
grid_r = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), Ridge(alpha=1)),
                    param_grid={'polynomialfeatures__degree': [i for i in range(5)],
                    'ridge__alpha': [10**i for i in range(-10,10)]},
                    cv=kfold,
                    refit=True)
```

```python
grid_linreg = GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), LinearRegression()),
                    param_grid={'polynomialfeatures__degree': [i for i in range(5)]},
                    cv=kfold,
                    refit=True)
```

```python
grid_eln.fit(X_train, y_train)
grid_ls.fit(X_train, y_train)
grid_r.fit(X_train, y_train)
grid_linreg.fit(X_train, y_train)
```

```python
from sklearn.svm import SVR

svr = GridSearchCV(SVR(kernel='rbf', gamma=0.1), cv=5,
                   param_grid={"C": [0.001, 0.01, 0.1, 1, 10 ,100, 1000],
                               "gamma": [0.001, 0.01, 0.1, 1, 10 ,100, 1000]})
```

```python
svr.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor

rfr = GridSearchCV(RandomForestRegressor(random_state=3), cv=5,
        param_grid={"max_depth": range(1,16),
        "n_estimators": range(1,16)},
        refit = True)
```

```python
rfr.fit(X_train,y_train)
```

```python
models.append(('ElasticNet', grid_eln.best_estimator_))
models.append(('Lasso', grid_ls.best_estimator_))
models.append(('Ridge', grid_r.best_estimator_))
models.append(('LR', grid_linreg.best_estimator_))
models.append(('SVR', svr.best_estimator_))
models.append(('RFR', rfr.best_estimator_))
```

```python
from sklearn import  metrics

r2 = []
explained_variance_score = []
median_absolute_error = []
mean_squared_error = []
mean_absolute_error = []
for name, model in models:
    explained_variance_score.append(metrics.explained_variance_score(y_test, model.predict(X_test)))
    mean_squared_error.append(metrics.mean_squared_error(y_test, model.predict(X_test)))
    mean_absolute_error.append(metrics.mean_absolute_error(y_test, model.predict(X_test)))
    median_absolute_error.append( metrics.median_absolute_error(y_test, model.predict(X_test)))
    r2.append(metrics.r2_score(y_test, model.predict(X_test)))
```

```python
import pandas as pd
d = {'explained_variance_score': explained_variance_score,
     'mean_squared_error' : mean_squared_error,
     'mean_absolute_error' : mean_absolute_error,
     'median_absolute_error': median_absolute_error,
     'r2': r2
    }
df = pd.DataFrame(data=d)
df.insert(loc=0, column='Method', value=['ElasticNet','Lasso','Ridge','LR', 'SVR', 'RFR'])
df
```
