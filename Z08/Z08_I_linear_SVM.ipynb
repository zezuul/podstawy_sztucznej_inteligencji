# Linear SVM Classification

SVM is effective in high dimensional spaces.The goal of the SVM algorithm is to create the best line or decision boundary that can segregate n-dimensional space into classes so that we can easily put the new data point in the correct category in the future. This best decision boundary is called a hyperplane. 

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=.1, random_state=42)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
C = [0.0001, 0.001, 0.01, 0.1, 1, 10, 100]

for val in C:
    mod = SVC(kernel = 'linear', C = val)
    mod.fit(X, y)
    plot_decision_regions(X, y, mod)
    plt.show()
```

```python
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
for val in C:
    clf = make_pipeline(StandardScaler(), SVC(kernel='linear', C = val))
    clf.fit(X, y)
    plot_decision_regions(X, y, clf)
    plt.show()
```
